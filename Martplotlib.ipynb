```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

np.random.seed(42)

data = {
    "employee": ["Alice", "Bob", "Charlie", "Diana", "Eve",
                 "Frank", "Grace", "Henry", "Isla", "Jack"],
    "department": ["IT", "HR", "IT", "Finance", "HR",
                   "IT", "Finance", "HR", "IT", "Finance"],
    "salary":     np.random.randint(40000, 120000, 10),
    "experience": np.random.randint(1, 15, 10),
    "rating":     np.random.randint(1, 10, 10)
}

df = pd.DataFrame(data)
new_df=df.groupby('department')['salary'].mean()

print(new_df)

plt.bar(new_df.index,new_df.values)
plt.xlabel('Department')
plt.ylabel('Salary')
plt.title('Comparision of Department and Salary')
plt.show()

my_colors=df['department'].map({
    "IT":"green",
    "Finance":"blue",
    "HR":"red"
})

plt.scatter(df['experience'],df['salary'],c=my_colors)
plt.show()

corr=df[['experience','salary','rating']].corr()
print(corr)


sns.boxplot(x=df['department'],y=df['salary'])
plt.show()


df.sort_values(by="experience",ascending=True,inplace=True)
plt.plot(df['experience'],df['salary'])
plt.show()
print(df)



```
