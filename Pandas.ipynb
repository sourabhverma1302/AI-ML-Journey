```python
import pandas as pd
import numpy as np

# Dataset with missing values
data = {
    "employee": ["Alice", "Bob", "Charlie", np.nan, "Eve", "Frank"],
    "salary":   [50000, np.nan, 75000, np.nan, 45000, 60000],
    "age":      [25, 30, np.nan, 28, np.nan, 35],
    "department": ["IT", "HR", np.nan, "IT", "Finance", np.nan]
}

df = pd.DataFrame(data)


df['salary'].fillna(df['salary'].mean(),inplace=True)
df['age'].fillna(df['age'].mean(),inplace=True)
df['department'].fillna('Unknown',inplace=True)
df.dropna(how='any',inplace=True)
print(df.isna().sum())

# Write code to:
# 1. Check how many missing values in each column
# 2. Fill missing salary with the average salary
# 3. Fill missing age with the median age
# 4. Fill missing department with "Unknown"
# 5. Drop any rows that still have missing values
# 6. Confirm no missing values remain
```

```python
# DataFrame 1 — employee info
employees = pd.DataFrame({
    "emp_id":   [1, 2, 3, 4, 5],
    "name":     ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "dept_id":  [101, 102, 101, 103, 102]
})

# DataFrame 2 — department info
departments = pd.DataFrame({
    "dept_id":  [101, 102, 103,104],
    "dept_name": ["IT", "HR", "Finance","IT"],
    "location":  ["Perth", "Sydney", "Melbourne","Brisbane"]
})

merged=pd.merge(employees,departments,on="dept_id",how="inner")
print(merged)

IT_employees=merged[merged['dept_name'] == 'IT']
print(IT_employees)


print("employeeName_City")

employeeName_City=merged[["name","location"]]
print(employeeName_City)




grouped=merged.groupby('dept_name').count()
print(grouped)



# Write code to:
# 1. Merge both DataFrames on dept_id
# 2. Show only employees in IT department
# 3. Show employee name + their city (location)
# 4. Count how many employees per department
```
